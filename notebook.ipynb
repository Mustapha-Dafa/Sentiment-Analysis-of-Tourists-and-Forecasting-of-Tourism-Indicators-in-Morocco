```python
"""
Analyse de Corrélation CORRIGÉE pour Google Trends × Tourisme Maroc
Résout les problèmes des résultats illogiques

PROBLÈMES IDENTIFIÉS:
1. Corrélations négatives pour keywords touristiques évidents (riad, ait benhaddou)
2. Domination de "argan oil" (pas directement lié au tourisme)
3. Lag uniformément à 10 mois (suspect)
4. Keywords non-touristiques dans les résultats (bbc good food)

SOLUTIONS APPLIQUÉES:
- Nettoyage strict des keywords non-touristiques
- Normalisation/détrending des séries temporelles
- Analyse de corrélation multiple (lag 0-12 mois)
- Validation croisée avec saisonnalité
- Détection d'outliers (COVID)
"""

import pandas as pd
import numpy as np
from pathlib import Path
from scipy import stats
from scipy.stats import pearsonr
import warnings
warnings.filterwarnings('ignore')

# ========= CONFIGURATION =========
TRENDS_FILE = Path("trends.csv")
ARRIVALS_FILE = Path("/data/tourism/KPI_touristique.csv")  # Format: Date, Arrivals
OUTPUT_DIR = Path("correlation_analysis")
OUTPUT_DIR.mkdir(exist_ok=True)

# Paramètres d'analyse
LAG_RANGE = range(1, 13)  # 0 à 12 mois
MIN_CORRELATION = 0.5      # Seuil minimum
COVID_START = "2020-03-01"  # Début période COVID
COVID_END = "2021-12-31"    # Fin période COVID

# ========= ÉTAPE 1: FILTRAGE STRICT DES KEYWORDS =========

def filter_tourism_keywords(keywords):
    """
    Filtre STRICT: garde uniquement les keywords touristiques pertinents
    Élimine: produits commerciaux, recettes, news, etc.
    """
    # Patterns NON-touristiques à exclure
    EXCLUDE_PATTERNS = [
        r'\b(oil|argan|beauty|cosmetic|shampoo|body|hair)\b',  # Produits beauté
        r'\b(recipe|food|chilli|chocolate|cake|cooking)\b',     # Recettes
        r'\b(bbc|news|article|blog|instagram|facebook)\b',      # Média/social
        r'\b(pcr|test|covid|vaccine|quarantine)\b',             # COVID (peut être gardé séparément)
        r'\b(price|buy|shop|store|amazon|ebay)\b',              # E-commerce
        r'\b(job|work|salary|visa application)\b',              # Immigration/travail
    ]
    
    # Patterns TOURISTIQUES à inclure
    INCLUDE_PATTERNS = [
        r'\b(hotel|riad|resort|accommodation|stay|booking)\b',
        r'\b(flight|airport|airline|ticket|travel)\b',
        r'\b(tour|excursion|visit|guide|itinerary|trip)\b',
        r'\b(marrakech|casablanca|fez|chefchaouen|essaouira|tangier|agadir)\b',
        r'\b(sahara|desert|atlas|medina|kasbah|souk)\b',
        r'\b(weather|temperature|climate|best time)\b',
        r'\b(things to do|places to visit|attractions)\b',
        r'\b(restaurant|food|cuisine|dining)\b',  # Gastronomie touristique OK
        r'\b(beach|mountain|hiking|surfing|activities)\b',
    ]
    
    import re
    filtered = []
    
    for kw in keywords:
        kw_lower = kw.lower()
        
        # Exclure si match pattern non-touristique
        if any(re.search(pattern, kw_lower) for pattern in EXCLUDE_PATTERNS):
            continue
        
        # Inclure si match pattern touristique
        if any(re.search(pattern, kw_lower) for pattern in INCLUDE_PATTERNS):
            filtered.append(kw)
    
    return filtered

# ========= ÉTAPE 2: PRÉPARATION DES DONNÉES =========

def load_and_prepare_data():
    """Charge et prépare les données trends et arrivées"""
    
    # 1. Charger trends
    if not TRENDS_FILE.exists():
        print(f"❌ Fichier {TRENDS_FILE} introuvable")
        return None, None
    
    trends = pd.read_csv(TRENDS_FILE, parse_dates=['Date'], index_col='Date')
    print(f"✓ Trends chargés: {len(trends)} dates × {len(trends.columns)} keywords")
    
    # 2. Charger arrivées touristiques
    if not ARRIVALS_FILE.exists():
        print(f"\n⚠️  Fichier arrivées touristiques manquant: {ARRIVALS_FILE}")
        print("   Création de données SIMULÉES pour démonstration...")
        arrivals = generate_synthetic_arrivals(trends.index)
    else:
        arrivals = pd.read_csv(ARRIVALS_FILE, parse_dates=['Date'], index_col='Date')
        arrivals = arrivals['TES']
        print(f"✓ Arrivées chargées: {len(arrivals)} dates")
    
    # 3. Aligner les dates (intersection)
    common_dates = trends.index.intersection(arrivals.index)
    trends = trends.loc[common_dates]
    arrivals = arrivals.loc[common_dates]
    
    print(f"✓ Période commune: {common_dates.min().date()} → {common_dates.max().date()}")
    
    return trends, arrivals

def generate_synthetic_arrivals(date_index):
    """
    Génère des données d'arrivées SIMULÉES avec saisonnalité réaliste
    (pour démonstration uniquement - remplacer par vraies données)
    """
    np.random.seed(42)
    
    arrivals = pd.Series(index=date_index, dtype=float)
    
    for date in date_index:
        # Saisonnalité: pic en été (juin-août), creux en hiver
        month = date.month
        seasonal_factor = 1.0 + 0.5 * np.sin((month - 3) * np.pi / 6)
        
        # Croissance tendancielle
        year = date.year
        trend_factor = 1.0 + (year - date_index.min().year) * 0.05
        
        # Choc COVID
        if COVID_START <= date.strftime("%Y-%m-%d") <= COVID_END:
            covid_factor = 0.2  # -80% pendant COVID
        else:
            covid_factor = 1.0
        
        # Valeur finale
        base_value = 100000  # 100k touristes/mois en base
        arrivals[date] = base_value * seasonal_factor * trend_factor * covid_factor
        arrivals[date] += np.random.normal(0, 5000)  # Bruit
    
    return arrivals

# ========= ÉTAPE 3: NETTOYAGE ET NORMALISATION =========

def clean_and_normalize(series, remove_covid=True):
    """
    Nettoie et normalise une série temporelle
    - Supprime outliers
    - Normalise (0-100)
    - Optionnel: exclut période COVID
    """
    series = series.copy()
    
    # Option: exclure période COVID
    if remove_covid:
        covid_mask = (series.index >= COVID_START) & (series.index <= COVID_END)
        series = series[~covid_mask]
    
    # Remplacer valeurs manquantes par interpolation
    series = series.interpolate(method='linear')
    
    # Normaliser (0-100)
    if series.std() > 0:
        series = (series - series.mean()) / series.std()
        series = (series - series.min()) / (series.max() - series.min()) * 100
    
    return series

# ========= ÉTAPE 4: ANALYSE DE CORRÉLATION MULTI-LAG =========

def calculate_lagged_correlation(keyword_series, arrivals_series, max_lag=12):
    """
    Calcule la corrélation pour différents lags (0-12 mois)
    Retourne le lag optimal et sa corrélation
    """
    results = []
    
    for lag in range(0, max_lag + 1):
        # Décaler la série keyword de 'lag' mois
        keyword_lagged = keyword_series.shift(lag)
        
        # Aligner les dates (supprimer NaN)
        aligned = pd.DataFrame({
            'keyword': keyword_lagged,
            'arrivals': arrivals_series
        }).dropna()
        
        if len(aligned) < 12:  # Minimum 12 mois de données
            continue
        
        # Calculer corrélation de Pearson
        corr, pvalue = pearsonr(aligned['keyword'], aligned['arrivals'])
        
        results.append({
            'lag': lag,
            'correlation': corr,
            'pvalue': pvalue,
            'n_observations': len(aligned)
        })
    
    if not results:
        return None
    
    # Sélectionner le lag avec la meilleure corrélation (en valeur absolue)
    best = max(results, key=lambda x: abs(x['correlation']))
    return best

# ========= ÉTAPE 5: ANALYSE PRINCIPALE =========

def perform_correlation_analysis(trends, arrivals):
    """Analyse de corrélation complète avec toutes les corrections"""
    
    print("\n" + "="*80)
    print("ANALYSE DE CORRÉLATION: GOOGLE TRENDS × ARRIVÉES TOURISTIQUES")
    print("="*80)
    
    # 1. Filtrage des keywords touristiques
    all_keywords = trends.columns.tolist()
    tourism_keywords = filter_tourism_keywords(all_keywords)
    
    print(f"\n📊 Filtrage des keywords:")
    print(f"   Total: {len(all_keywords)}")
    print(f"   Touristiques: {len(tourism_keywords)}")
    print(f"   Exclus: {len(all_keywords) - len(tourism_keywords)}")
    
    # 2. Normaliser les arrivées
    arrivals_clean = clean_and_normalize(arrivals, remove_covid=False)
    
    # 3. Analyse par keyword
    results = []
    
    print(f"\n🔄 Analyse en cours...")
    for i, keyword in enumerate(tourism_keywords, 1):
        if i % 10 == 0:
            print(f"   Progression: {i}/{len(tourism_keywords)}")
        
        keyword_series = trends[keyword]
        keyword_clean = clean_and_normalize(keyword_series, remove_covid=False)
        
        # Calculer corrélation multi-lag
        best_lag = calculate_lagged_correlation(keyword_clean, arrivals_clean)
        
        if best_lag and abs(best_lag['correlation']) >= MIN_CORRELATION:
            results.append({
                'keyword': keyword,
                'best_lag': best_lag['lag'],
                'correlation': best_lag['correlation'],
                'pvalue': best_lag['pvalue'],
                'n_obs': best_lag['n_observations'],
                'significant': best_lag['pvalue'] < 0.05
            })
    
    # 4. Trier par corrélation (valeur absolue décroissante)
    results_df = pd.DataFrame(results)
    results_df = results_df.sort_values('correlation', key=abs, ascending=False)
    
    return results_df

# ========= ÉTAPE 6: RAPPORT ET EXPORT =========

def generate_report(results):
    """Génère un rapport d'analyse détaillé"""
    
    print("\n" + "="*80)
    print("RÉSULTATS - TOP KEYWORDS PAR CORRÉLATION")
    print("="*80)
    
    if len(results) == 0:
        print("\n⚠️  Aucun keyword ne dépasse le seuil de corrélation")
        return
    
    # Séparer corrélations positives et négatives
    positive = results[results['correlation'] > 0]
    negative = results[results['correlation'] < 0]
    
    print(f"\n✅ CORRÉLATIONS POSITIVES (indicateurs prédictifs): {len(positive)}")
    print("-"*80)
    if len(positive) > 0:
        print(positive[['keyword', 'best_lag', 'correlation', 'significant']].head(15).to_string(index=False))
    
    print(f"\n❌ CORRÉLATIONS NÉGATIVES (anti-indicateurs): {len(negative)}")
    print("-"*80)
    if len(negative) > 0:
        print(negative[['keyword', 'best_lag', 'correlation', 'significant']].head(10).to_string(index=False))
    
    # Statistiques par lag
    print(f"\n📊 DISTRIBUTION DES LAGS:")
    print("-"*80)
    lag_dist = results['best_lag'].value_counts().sort_index()
    for lag, count in lag_dist.items():
        print(f"   Lag {lag:2d} mois: {count:3d} keywords")
    
    # Sauvegarde
    output_file = OUTPUT_DIR / "correlation_results.csv"
    results.to_csv(output_file, index=False)
    print(f"\n💾 Résultats sauvegardés: {output_file}")
    
    # Recommandations
    print("\n" + "="*80)
    print("💡 RECOMMANDATIONS")
    print("="*80)
    
    if len(positive) > 0:
        avg_lag = positive['best_lag'].mean()
        print(f"✓ Lag moyen optimal: {avg_lag:.1f} mois")
        print(f"✓ Planifier campagnes marketing {int(avg_lag)} mois avant la haute saison")
        print(f"✓ Top 3 indicateurs prédictifs:")
        for idx, row in positive.head(3).iterrows():
            print(f"   • {row['keyword'][:50]:50s} (lag={row['best_lag']}, r={row['correlation']:.3f})")
    
    if len(negative) > 0:
        print(f"\n⚠️  {len(negative)} keywords ont une corrélation négative")
        print(f"   → Possibles causes: recherches post-crise, confusion sémantique")
        print(f"   → Investiguer manuellement ces keywords")

# ========= FONCTION PRINCIPALE =========

def main():
    """Pipeline complet d'analyse"""
    
    print("\n" + "="*80)
    print("ANALYSE DE CORRÉLATION CORRIGÉE - TOURISME MAROC")
    print("="*80)
    
    # 1. Charger données
    trends, arrivals = load_and_prepare_data()
    if trends is None:
        return
    
    # 2. Analyse de corrélation
    results = perform_correlation_analysis(trends, arrivals)
    
    # 3. Rapport
    generate_report(results)
    
    print("\n✅ ANALYSE TERMINÉE")
    print("="*80)

if __name__ == "__main__":
    main()
```

```python
"""
Analyse de Corrélation SIMPLIFIÉE : Google Trends × Tourisme Maroc
- Pas de données synthétiques
- Pas de standardisation
- Code minimaliste
"""

import pandas as pd
from pathlib import Path
from scipy.stats import pearsonr
import warnings
warnings.filterwarnings('ignore')

# ========= CONFIGURATION =========
TRENDS_FILE = Path("trends.csv")
ARRIVALS_FILE = Path("data/tourism/KPI_touristique.csv")
OUTPUT_DIR = Path("correlation_results")
OUTPUT_DIR.mkdir(exist_ok=True)

# Paramètres
MAX_LAG = 12           # Tester jusqu'à 12 mois de décalage
MIN_CORRELATION = 0.5  # Seuil minimum
MIN_PVALUE = 0.05      # Significativité statistique

# ========= FILTRAGE DES KEYWORDS TOURISTIQUES =========

def filter_tourism_keywords(keywords):
    """Garde uniquement les mots-clés touristiques pertinents"""
    import re
    
    # Mots-clés à EXCLURE (non-touristiques)
    EXCLUDE = [
        r'\b(oil|argan|beauty|cosmetic|shampoo|body|hair)\b',
        r'\b(recipe|food|chilli|chocolate|cake|cooking)\b',
        r'\b(bbc|news|article|blog|instagram|facebook)\b',
        r'\b(price|buy|shop|store|amazon|ebay)\b',
        r'\b(job|work|salary|visa application)\b',
    ]
    
    # Mots-clés à INCLURE (touristiques)
    INCLUDE = [
        r'\b(hotel|riad|resort|accommodation|stay|booking)\b',
        r'\b(flight|airport|airline|ticket|travel)\b',
        r'\b(tour|excursion|visit|guide|itinerary|trip)\b',
        r'\b(marrakech|casablanca|fez|chefchaouen|essaouira|tangier|agadir)\b',
        r'\b(sahara|desert|atlas|medina|kasbah|souk)\b',
        r'\b(weather|temperature|climate|best time)\b',
        r'\b(things to do|places to visit|attractions)\b',
        r'\b(restaurant|cuisine|dining)\b',
        r'\b(beach|mountain|hiking|surfing|activities)\b',
    ]
    
    filtered = []
    for kw in keywords:
        kw_lower = kw.lower()
        
        # Exclure si non-touristique
        if any(re.search(pattern, kw_lower) for pattern in EXCLUDE):
            continue
        
        # Inclure si touristique
        if any(re.search(pattern, kw_lower) for pattern in INCLUDE):
            filtered.append(kw)
    
    return filtered

# ========= CHARGEMENT DES DONNÉES =========

def load_data():
    """Charge les fichiers CSV"""
    
    # Vérifier que les fichiers existent
    if not TRENDS_FILE.exists():
        raise FileNotFoundError(f"❌ Fichier manquant : {TRENDS_FILE}")
    
    if not ARRIVALS_FILE.exists():
        raise FileNotFoundError(f"❌ Fichier manquant : {ARRIVALS_FILE}")
    
    # Charger Google Trends
    trends = pd.read_csv(TRENDS_FILE, parse_dates=['Date'], index_col='Date')
    print(f"✓ Trends : {len(trends)} lignes × {len(trends.columns)} mots-clés")
    
    # Charger arrivées touristiques
    arrivals = pd.read_csv(ARRIVALS_FILE, parse_dates=['Date'], index_col='Date')
    arrivals = arrivals['TES']  # Colonne des arrivées
    print(f"✓ Arrivées : {len(arrivals)} lignes")
    
    # Aligner les dates (garder uniquement les dates communes)
    common_dates = trends.index.intersection(arrivals.index)
    trends = trends.loc[common_dates]
    arrivals = arrivals.loc[common_dates]
    
    print(f"✓ Période : {common_dates.min().date()} → {common_dates.max().date()}")
    
    return trends, arrivals

# ========= CALCUL DE CORRÉLATION AVEC LAG =========

def calculate_correlation_with_lag(keyword_series, arrivals_series, max_lag=12):
    """
    Teste différents lags (0 à max_lag mois)
    Retourne le lag avec la meilleure corrélation
    """
    results = []
    
    for lag in range(1, max_lag + 1):
        # Décaler la série keyword de 'lag' mois
        keyword_lagged = keyword_series.shift(lag)
        
        # Créer un DataFrame aligné
        df = pd.DataFrame({
            'keyword': keyword_lagged,
            'arrivals': arrivals_series
        }).dropna()  # Supprimer les valeurs manquantes
        
        # Minimum 12 observations pour une corrélation fiable
        if len(df) < 12:
            continue
        
        # Calculer la corrélation de Pearson
        corr, pvalue = pearsonr(df['keyword'], df['arrivals'])
        
        results.append({
            'lag': lag,
            'correlation': corr,
            'pvalue': pvalue,
            'n_observations': len(df)
        })
    
    if not results:
        return None
    
    # Retourner le lag avec la plus forte corrélation (en valeur absolue)
    best = max(results, key=lambda x: abs(x['correlation']))
    return best

# ========= ANALYSE PRINCIPALE =========

def analyze_correlations(trends, arrivals):
    """Analyse de corrélation pour tous les mots-clés"""
    
    print("\n" + "="*80)
    print("ANALYSE DE CORRÉLATION")
    print("="*80)
    
    # Filtrer les mots-clés touristiques
    all_keywords = trends.columns.tolist()
    tourism_keywords = filter_tourism_keywords(all_keywords)
    
    print(f"\n📊 Mots-clés : {len(tourism_keywords)} touristiques / {len(all_keywords)} total")
    
    # Analyser chaque mot-clé
    results = []
    
    print(f"\n🔄 Analyse en cours...")
    for i, keyword in enumerate(tourism_keywords, 1):
        if i % 20 == 0:
            print(f"   {i}/{len(tourism_keywords)}")
        
        # Calculer la meilleure corrélation avec lag
        best = calculate_correlation_with_lag(trends[keyword], arrivals, MAX_LAG)
        
        # Garder si corrélation suffisante et significative
        if best and abs(best['correlation']) >= MIN_CORRELATION and best['pvalue'] < MIN_PVALUE:
            results.append({
                'keyword': keyword,
                'lag_mois': best['lag'],
                'correlation': round(best['correlation'], 3),
                'pvalue': best['pvalue'],
                'n_obs': best['n_observations']
            })
    
    # Créer DataFrame et trier par corrélation
    df_results = pd.DataFrame(results)
    df_results = df_results.sort_values('correlation', key=abs, ascending=False)
    
    return df_results

# ========= AFFICHAGE DES RÉSULTATS =========

def display_results(results):
    """Affiche et sauvegarde les résultats"""
    
    print("\n" + "="*80)
    print("RÉSULTATS")
    print("="*80)
    
    if len(results) == 0:
        print("\n⚠️  Aucun mot-clé ne dépasse les seuils")
        return
    
    # Séparer positives et négatives
    positive = results[results['correlation'] > 0]
    negative = results[results['correlation'] < 0]
    
    # Afficher corrélations positives
    print(f"\n✅ CORRÉLATIONS POSITIVES : {len(positive)}")
    print("-"*80)
    if len(positive) > 0:
        print(positive[['keyword', 'lag_mois', 'correlation']].head(20).to_string(index=False))
    
    # Afficher corrélations négatives
    if len(negative) > 0:
        print(f"\n❌ CORRÉLATIONS NÉGATIVES : {len(negative)}")
        print("-"*80)
        print(negative[['keyword', 'lag_mois', 'correlation']].head(10).to_string(index=False))
    
    # Distribution des lags
    print(f"\n📊 DISTRIBUTION DES LAGS :")
    print("-"*80)
    lag_counts = results['lag_mois'].value_counts().sort_index()
    for lag, count in lag_counts.items():
        print(f"   Lag {lag:2d} mois : {count:3d} mots-clés")
    
    # Sauvegarder en CSV
    output_file = OUTPUT_DIR / "correlation_results.csv"
    results.to_csv(output_file, index=False)
    print(f"\n💾 Sauvegardé : {output_file}")
    
    # Recommandations
    if len(positive) > 0:
        print("\n" + "="*80)
        print("💡 RECOMMANDATIONS")
        print("="*80)
        avg_lag = positive['lag_mois'].mean()
        print(f"✓ Lag moyen : {avg_lag:.1f} mois")
        print(f"✓ Top 3 indicateurs prédictifs :")
        for idx, row in positive.head(3).iterrows():
            print(f"   • {row['keyword']} (lag={row['lag_mois']}, r={row['correlation']})")

# ========= EXÉCUTION =========

def main():
    """Fonction principale"""
    
    print("\n" + "="*80)
    print("ANALYSE CORRÉLATION - GOOGLE TRENDS × TOURISME MAROC")
    print("="*80)
    
    try:
        # 1. Charger les données
        trends, arrivals = load_data()
        
        # 2. Analyser les corrélations
        results = analyze_correlations(trends, arrivals)
        
        # 3. Afficher les résultats
        display_results(results)
        
        print("\n✅ TERMINÉ")
        print("="*80)
        
    except FileNotFoundError as e:
        print(f"\n{e}")
        print("\nAssurez-vous que les fichiers suivants existent :")
        print(f"   - {TRENDS_FILE}")
        print(f"   - {ARRIVALS_FILE}")
    except Exception as e:
        print(f"\n❌ Erreur : {e}")

if __name__ == "__main__":
    main()
```
