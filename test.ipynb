```python
import pandas as pd
```

```python
df=pd.read_csv("./data/Google_reviews.csv")
```

```python
df[df["city"]=="Agadir"]["date"]
```

```python
def par_moi(text):
    x=text.split("/")[1]
    y=text.split("/")[2]
    return f"{x}/{y}"
```

```python
df["date_par_mois"]=df["date"].apply(par_moi)
```

```python
df[df["city"]=="Agadir"]["date_par_mois"].unique().shape[0]
```

```python
cities=df["city"].unique()
```

```python
for cit in cities:
    x=df[df["city"]==cit]["date_par_mois"].unique()
    print(f"  la ville {cit} est: {x}")
    print()
```
