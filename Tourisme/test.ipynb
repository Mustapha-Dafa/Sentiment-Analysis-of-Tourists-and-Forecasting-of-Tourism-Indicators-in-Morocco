```python
annees=[2012,2013,2015,2016]
annees.reverse()
```

```python

taux_par_annee = {an: globals()[f"taux_{an}"] 
                 for an in annees 
                 if f"taux_{an}" in globals()}


nuits=[]
for i in range(len(annees)):
    for j in range(12):
        mois=f"0{j+1}" if j<9 else f"{j+1}"
        nuits.append({
            "Date":f"{mois}/{annees[i]}",
            "Taux d'occupation":taux_par_annee[annees[i]][j]
        })
    
```

```python
tes=pd.DataFrame(nuits)
```

```python
dtt=pd.concat([dr,tes],ignore_index=True)
dtt.to_csv("Taux_occupation.csv",index=False)
```
