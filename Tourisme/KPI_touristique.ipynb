```python
data1=pd.read_csv("./data/nuites_par_date_categorie_2014.csv")
data1=data1.drop(["date"],axis=1)
data1["mm_yy"]
```

```python
def change(text):
    text=str(text)
    if "14" in text:
        return text.replace("14","15")
```

```python
data1["mm_yy"]=data1["mm_yy"].apply(change)
data1
```
